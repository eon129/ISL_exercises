### Error distribution analisis

#### Check y distribution when error is normally distributed

```python
#Number of data points
n = 200
X = np.random.uniform(-10,10, n) 
error = np.random.normal(0, 30, n) # Gaussian distribution

#This are the real parameters
true_beta1, true_beta0 = 6, 2

#Our true function
# y = beta1 * x + beta0 + ε
y = true_beta1 * X + true_beta0 + error

pop = np.array([X, y])

plt.plot(X, y, 'x');
plt.xlabel('Example population')
plt.show()
```

```python
#plot error
plt.hist(error, bins=70)
plt.show()
```

```python
#plot y
plt.hist(y, bins=70)
plt.show()
```

#### Check with error binomial distributed

```python
#Number of data points
n = 200
X = np.random.uniform(-10,10, n) 
error = np.random.binomial(n, 0.1, n) # binomial distribution

#This are the real parameters
true_beta1, true_beta0 = 6, 2

#Our true function
# y = beta1 * x + beta0 + ε
y = true_beta1 * X + true_beta0 + error

pop = np.array([X, y])

plt.plot(X, y, 'x');
plt.xlabel('Example population')
plt.show()
```
